```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv("D:\\projects\\sales data.csv")
```

```python
#Created this method to figure out how many rows had duplicates and missing values(Na)
def analyze_missing_and_duplicates(df):
    print("Missing values per column:")
    print(df.isnull().sum())
    print("\nNumber of duplicate rows:")
    print(df.duplicated().sum())


#created this method for missing zipcodes (vermont zipcode: 5401)
def fill_missing_postal_code(df, state, postal_code):
    missing_values = df[df['Postal Code'].isnull()]
    print("\nRows with missing Postal Code:")
    print(missing_values)
    df.loc[(df['State'] == state) & (df['Postal Code'].isna()), 'Postal Code'] = postal_code
    print(f"\nAfter filling missing Postal Code for {state}:")
    print(df[df['State'] == state]['Postal Code'].unique())


#Most dates were invalid and we switched them to datetime data type
def handle_date_columns(df, date_columns, date_format='%d/%m/%Y'):
    for column in date_columns:
        df[column] = pd.to_datetime(df[column], format=date_format, errors='coerce')
        invalid_dates = df[df[column].isna()]
        print(f"\nInvalid dates in {column}:")
        print(invalid_dates)
    df_cleaned = df.dropna(subset=date_columns).copy()
    return df_cleaned


# IQR = q3-q1  https://youtu.be/A3gClkblXK8?si=BrwpqnT5M49Rpa0z
def remove_outliers(df, column):
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1-1.5*iqr
    upper_bound = q3+1.5*iqr
    print(f"\nOutlier thresholds for '{column}':")
    print(f"Lower bound: {lower_bound}, Upper bound: {upper_bound}")
    df_outliers_removed = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)].copy()
    print(f"Outliers removed: {len(df)-len(df_outliers_removed)}")
    return df_outliers_removed

```

```python
#To check that the data is cleaned properly
analyze_missing_and_duplicates(data)
fill_missing_postal_code(data, state='Vermont', postal_code=5401)
```

```python
date_columns = ['Order Date', 'Ship Date']
data_cleaned = handle_date_columns(data, date_columns)
data_cleaned = remove_outliers(data_cleaned, column='Profit')
```

```python
print("\nMissing values after cleaning:")
print(data_cleaned.isnull().sum())
print("\nShape of cleaned dataset:")
print(data_cleaned.shape)
```
